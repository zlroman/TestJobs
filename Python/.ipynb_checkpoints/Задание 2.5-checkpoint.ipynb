```python
Seasons = pd.DataFrame([[1, 'spring'], [2, 'summer'], [3, 'fall'], [4, 'winter']], columns=['season_id', 'season_name']) #задаем датафрейм
```

```python
bike1 = pd.merge( #данная функция merge соединяет два датафрейма в один по нами определенному правилу.
    bike,
    Seasons,
    left_on='season',
    right_on='season_id'
)
```

```python
bike1[['season_name', 'count']].groupby('season_name').count()
```

```python
pt = pd.pivot_table(bike, values='count', index=['season'], aggfunc=sum)
pt.div(pt.sum(0))
```
